```python
import pandas as pd    #Importing required libraries
import numpy as np
```

```python
df1=pd.read_csv('actual33.csv') # storing our actual and predicted qrels in two diff dataframes
df2=pd.read_csv('predictedquery3.csv')
```

```python

df1.head()
```

```python
df11 =df1.sort_values(by=['docid']) #sorting values
df11
```

```python
df22 =df2.sort_values(by=['PMCID'])
df22
```

```python
matchingdocs = df11.merge(df22, left_on='docid', right_on='PMCID') # checking the matching of retrieved document ids in actual and predicted
```

```python
matchingdocs
```

```python
TP = matchingdocs.shape[0] # True postive
```

```python
total = df22.shape[0]
```

```python
precision = TP/total # Precision calculation 
```

```python
## This gives us the percentage of the relevant documents retrieved in the top n results.
```

```python
precision   
```

```python
recall = TP/df11.shape[0] # Recall calculation
```
